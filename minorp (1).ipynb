```python
import numpy as np
import pandas as pd
import re
from collections import Counter
from datetime import datetime
```

```python
with open("hostel_bois.txt", "r", encoding="utf-8") as file:
  chat = file.readlines()

  print("Total Lines :", len(chat))
```

```python
chat = [line.strip() for line in chat if line.strip()]

print("Lines after cleaning :", len(chat))
```

```python
dates =[]
times = []
senders = []
messages = []
```

```python
dates = []
times = []
senders = []
messages = []

for line in chat:
    if " - " in line and ": " in line:
        parts = line.split(" - ", 1)
        if len(parts) == 2:
            date_time = parts[0]
            message_part = parts[1]

            dt = date_time.split(", ", 1)
            if len(dt) == 2 and ": " in message_part:
                date = dt[0]
                time = dt[1]

                sender, message = message_part.split(": ", 1)

                dates.append(date)
                times.append(time)
                senders.append(sender)
                messages.append(message)

print("Messages Parsed:", len(messages))
```

```python
# create Data frame

df = pd.DataFrame({
    "Date": dates,
    "Time": times,
    "Sender": senders,
    "Message": messages
})
df.head()
```

```python
from collections import Counter

words = []

for msg in df["Message"]:
    words.extend(msg.lower().split())

word_count = Counter(words)

print("Top 10 Most Used Words:")
print(word_count.most_common(10))
```
